# Question 1: Linear Regression using Gradient Descent

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

data = pd.read_csv("Training_set_heights200.csv")

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)

X = scaled_data[:, 1].reshape(-1, 1) 
y = scaled_data[:, 0] 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state=0)
def gradient_descent(X, y, m, n, learning_rate):

    N = len(X)

    y_pred = m * X + n

    dm = (-2/N) * sum(X * (y - y_pred))
    dn = (-2/N) * sum(y - y_pred)

    m_new = m - learning_rate * dm
    n_new = n - learning_rate * dn
    return m_new, n_new
def train_model(X, y, learning_rate=0.1, max_iterations=10000):
    m = 0
    n = 0
    m_history = [0, 0]
    n_history = [0, 0]
    for iteration in range(max_iterations):
# Get new parameters
        m_new, n_new = gradient_descent(X.flatten(), y, m, n, learning_rate)
# Update history
        m_history = [m_history[1], m_new]
        n_history = [n_history[1], n_new]
# Check convergence to 4 decimal places
        if (round(m_history[0], 4) == round(m_history[1], 4) and
            round(n_history[0], 4) == round(n_history[1], 4)):
            print(f"Converged after {iteration + 1} iterations")
            return m_new, n_new
        m, n = m_new, n_new
# Print progress every 1000 iterations
        if iteration % 1000 == 0:
            print(f"Iteration {iteration}: m = {m:.6f}, n = {n:.6f}")
    print("Maximum iterations reached")
    return m, n
# Train the model
learning_rate = 0.1
m, n = train_model(X_train, y_train, learning_rate=learning_rate)
# Make predictions
y_train_pred = m * X_train.flatten() + n
y_test_pred = m * X_test.flatten() + n
train_error = mean_squared_error(y_train, y_train_pred)
test_error = mean_squared_error(y_test, y_test_pred)
print("\nFinal Parameters:")
print(f"m (slope): {m:.6f}")
print(f"n (intercept): {n:.6f}")
print(f"\nTraining Mean Squared Error: {train_error:.6f}")
print(f"Testing Mean Squared Error: {test_error:.6f}")
# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.plot(X_train, y_train_pred, color='red', label='Regression Line')
plt.xlabel('Scaled Weight')
plt.ylabel('Scaled Height')
plt.title('Height vs Weight Linear Regression')
plt.legend()
plt.show()
# Inverse transform predictions to original scale for interpretation
# Create sample points for the line in original scale
X_line = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_line = m * X_line.flatten() + n
# Combine features and predictions for inverse transform
X_line_original = scaler.inverse_transform(np.column_stack([y_line, X_line.flatten()]))
data_original = scaler.inverse_transform(np.column_stack([y, X.flatten()]))
plt.figure(figsize=(10, 6))
plt.scatter(data_original[:, 1], data_original[:, 0], color='blue',label='Actual Data')
plt.plot(X_line_original[:, 1], X_line_original[:, 0], color='red',label='Regression Line')
plt.xlabel('Weight (original scale)')
plt.ylabel('Height (original scale)')
plt.title('Height vs Weight Linear Regression (Original Scale)')
plt.legend()
plt.show()
```
