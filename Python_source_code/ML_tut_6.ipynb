```python
from sklearn import preprocessing
import numpy as np
import pandas as pd
df=pd.read_csv("liver_patient.csv")
df
```

```python
DF=df.drop(["Age","Gender"],axis=1)
DF
```

```python
#Min-Max scaling
f=preprocessing.MinMaxScaler()
x=f.fit_transform(DF)
print(x)
```

```python
#train test split of the data with test size 20%
X=x[:,0:8]
Y=x[:,8]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x[:,0:8],x[:,8],test_size
=0.2,random_state=0)
from sklearn.linear_model import LogisticRegression
logisticR = LogisticRegression()
logisticR.fit(X_train, y_train)
```

```python
y_pred=logisticR.predict(X_test)
print(y_pred)
from sklearn.metrics import accuracy_score
print("Model accuracy : ", accuracy_score(y_test,y_pred))
```

```python
from sklearn.model_selection import KFold, cross_val_score
logisticR = LogisticRegression()
kfold_validation=KFold(n_splits=5, shuffle=True,random_state=42)
results = cross_val_score(logisticR,X_train,y_train,scoring='accuracy', cv=kfold_validation)
```

```python
print("Cross-validation accuracy scores:", results)
print("Mean Accuracy:", np.mean(results))
```

```python
#Stratified K fold cross validation
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(logisticR,X_train,y_train, scoring='accuracy',cv=skf)
```

```python
print("Stratified CV Scores:", scores)
print("Mean accuracy:", scores.mean())
```

```python
# Grid search for Decision tree hyperparameter tuning
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
data = pd.read_csv('liver_patient.csv')
data.drop('Gender', axis=1, inplace=True)
data
```

```python
scaler = MinMaxScaler()
x = scaler.fit_transform(data)
X = x[:, 0:9]
Y = x[:, 9]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
param_grid = {
 'max_depth': [3, 5, 10],
 'min_samples_leaf': [1, 5, 10, 20]
}
dt_model = DecisionTreeClassifier(random_state=42)
grid_search = GridSearchCV(estimator=dt_model, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, Y_train)
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
Y_pred = best_model.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
print("Best Parameters:", best_params)
print("Best Cross-validation Accuracy:", grid_search.best_score_)
print("Test Accuracy:", accuracy)
```

```python
#ROC and AUC Curve
from sklearn import preprocessing
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
df=pd.read_csv("liver_patient.csv")
df
```

```python
DF=df.drop(["Age","Gender"],axis=1)
DF
```

```python
#Min-Max scaling
f=preprocessing.MinMaxScaler()
x=f.fit_transform(DF)
print(x)
```

```python
#train test split of the data with test size 20%
X=x[:,0:8]
Y=x[:,8]
X_train,X_test,y_train,y_test=train_test_split(x[:,0:8],x[:,8],test_size
=0.2,random_state=0)
logisticR = LogisticRegression()
logisticR.fit(X_train, y_train)
y_pred=logisticR.predict(X_test)
print(y_pred)
```

```python
fpr, tpr, _ = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)
print(f"AUC Score: {roc_auc:.2f}")
# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color="blue", label=f"ROC curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], color="gray", linestyle="--")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate (FPR)")
plt.ylabel("True Positive Rate (TPR)")
plt.title("ROC Curve")
plt.legend(loc="lower right")
plt.show()
```
