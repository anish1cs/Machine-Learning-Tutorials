```python
import pandas as pd
from sklearn.model_selection import train_test_split
df =pd.DataFrame(x)
X =df.drop('LotArea',axis=1)
X
```

```python
y = df['LotArea']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=42)
```
